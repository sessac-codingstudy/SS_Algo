### 두번째 문제
*문제 설명*  
  
정수 배열 nums가 주어질 때,
서로 다른 3개의 숫자를 선택하여 더한 값이 소수인 경우의 개수를 구하시오.

```python
from itertools import combinations
from math import isqrt

def solution(nums):
    def is_prime(x):
        if x < 2:
            return False
        for i in range(2, isqrt(x) + 1):
            if x % i == 0:
                return False
        return True
    
    count = 0
    
    # 서로 다른 3개 조합
    for comb in combinations(nums, 3):
        if is_prime(sum(comb)):
            count += 1
    
    return count
```

```python
solution([1, 2, 3, 4])
```

### 세번째 문제
*문제 설명*  
다음과 같은 값이 주어진다.  

- order : 각 메뉴별 주문 수량
- single_price : 각 메뉴의 단품 가격
- set_menu : 세트메뉴 구성 (각 세트가 포함하는 메뉴 수량 리스트)
- set_price : 각 세트의 가격
  
단품과 세트를 적절히 조합하여 전체 주문을 가장 저렴하게 구매할 때의 최소 비용을 구하라

```python
from functools import lru_cache

def solution(order, single_price, set_menu, set_price):
    n = len(order)
    offers = list(zip(set_menu, set_price))

    @lru_cache(None)
    def dfs(need):
        # need: tuple of remaining quantities
        # 기본값: 전부 단품 구매
        best = sum(need[i] * single_price[i] for i in range(n))

        # 세트 적용해보기
        for items, price in offers:
            # 이 세트를 적용할 수 있는지(초과 구매 방지)
            ok = True
            new_need = list(need)
            for i in range(n):
                if items[i] > new_need[i]:
                    ok = False
                    break
                new_need[i] -= items[i]
            if not ok:
                continue

            best = min(best, price + dfs(tuple(new_need)))

        return best

    return dfs(tuple(order))
```
