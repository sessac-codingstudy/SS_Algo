**문제**  
n개의 원소가 주어졌을 때, 처음에는 오름차순으로 정렬하여 출력하고, 그 다음에는 내림차순으로 정렬하여 출력하는 프로그램을 작성해보세요.  

**입력**  
첫번째 줄에 정수 n이 주어집니다.  
두번째 줄에는 n개의 원소가 공백을 사이에 두고 주어집니다.  

**출력**  
첫번째 줄에는 n개의 원소를 오름차순으로 정렬한 이후의 결과를 공백을 사이에 두고 출력합니다.  
두번째 줄에는 n개의 원소를 내림차순으로 청렬한 이후의 결과를 공백을 사엥 두고 출력합니다.

```python
n = int(input())
nums = list(map(int, input().split()))

# Please write your code here.
nums.sort()
print(*nums)

nums.sort(reverse=True)
print(' '.join(map(str, nums)))
```
