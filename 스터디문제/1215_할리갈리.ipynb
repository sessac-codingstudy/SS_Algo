**문제 요약**  
펼쳐진 카드 N장이 주어진다.  
각 카드는 `(과일종류 S, 개수 X)` 형태이며 과일 종류는 4개 중 하나이다.

- STRAWBERRY
- BANANA
- LIME
- PLUM

**펼쳐진 카드들 전체에서 어떤 과일이든 합이 정확히 5개가 되는 과일이 하나라도 있으면** 종을 쳐야 한다.

**출력**
- 종을 쳐야 하면 `YES`
- 아니면 `NO`

```python
n = int(input())

fruits = {
    "STRAWBERRY": 0,
    "BANANA": 0,
    "LIME": 0,
    "PLUM": 0
}

for _ in range(n):
    s, x = input().split()
    fruits[s] += int(x)


print("YES" if any(v == 5 for v in fruits.values()) else "NO")
```
