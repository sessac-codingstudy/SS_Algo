# 전주 듣고 노래 맞히기

## 문제 요약
- 정환이 **알고 있는 노래 N개**가 주어진다.
- 각 노래는
  - 제목 `S`
  - 처음 등장하는 7개의 음 `a1..a7`
  를 가진다.
- 정환은 **첫 3음(b1,b2,b3)**만 듣고 노래를 맞히려고 한다.
- 규칙:
  - 저장된 노래 중 **첫 3음이 같은 노래가 정확히 1개** → 그 노래 제목 출력
  - **2개 이상** → `?`
  - **0개** → `!`
- 대소문자 구분(제목은 그대로 출력)

```python
a, b = map(int, input().split())

music = {}
for _ in range(a):
    n = list(input().split())
    key = tuple(n[2:5])

    if key not in music:
        music[key] = n[1]
    else:
        music[key] = "?"


for _ in range(b):
    oppinions = tuple(input().split())
    if oppinions not in music:
        print("!")
    else:
        print(music[oppinions])
```
