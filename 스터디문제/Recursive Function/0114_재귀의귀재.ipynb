# 재귀의 귀재 백준 25501
https://www.acmicpc.net/problem/25501

```python
def recursion(s, l, r, count):
    if l >= r: 
        return 1, count
    elif s[l] != s[r]: 
        return 0, count
    else: 
        count += 1
        return recursion(s, l+1, r-1, count)

def isPalindrome(s):
    count = 1
    return recursion(s, 0, len(s)-1, count)

n = int(input())
for _ in range(n):
    A = input()
    print(*isPalindrome(A))
```

```python
print('ABBA:', isPalindrome('ABBA')[0], isPalindrome('ABBA')[1])
print('ABC:', *isPalindrome('ABC'))
```
