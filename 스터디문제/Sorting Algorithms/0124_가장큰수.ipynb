# 가장 큰 수
https://school.programmers.co.kr/learn/courses/30/lessons/42746

```python
def solution(citations):
    answer = 0
    citations.sort(reverse=True)
    
    for i in range(len(citations)):
        if citations[i] >= i+1:
            answer = i+1
        else:
            break
    return answer
```

```python
print(solution([3, 0, 6, 1, 5]))  # 3
```

## 해결 과정
1. 인용 횟수 배열을 내림차순으로 정렬한다.
2. 앞에서부터 하나씩 순회하며  
   `인용 횟수 ≥ 논문 수` 조건을 확인한다.
3. 조건을 만족한 최대 논문 수를 H-Index로 반환한다.
