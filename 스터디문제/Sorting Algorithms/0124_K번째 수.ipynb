# K번째 수 정렬
https://school.programmers.co.kr/learn/courses/30/lessons/42748?language=python3

```python
array = [1, 5, 2, 6, 3, 7, 4]
commands = [[2, 5, 3], [4, 4, 1], [1, 7, 3]]
```

```python
def solution(array, commands):
    answer = []
    for c in commands:
        i = int(c[0])
        j = int(c[1])
        k = int(c[2])
        
        aa = sorted(array[i-1:j])
        answer.append(aa[k-1])
    return answer
```

```python
solution(array, commands)
```
