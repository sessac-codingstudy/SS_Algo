````python
a = ['119','97674223','1195524421']
b = ['123','456','789']
c = ['12','123','1235','567','88']
d = ["123", "12", "456"]
````

````python
def solution_a(phone_book):
    answer = True
    for k in phone_book:
        result = ''
        for a in k:
            result += a
            if result != k and result in phone_book:
                answer = False
                break
    
    return answer
````

````python
def solution(phone_book):
    phone_book.sort()  # ÏÇ¨Ï†ÑÏàú Ï†ïÎ†¨

    for i in range(len(phone_book) - 1):
        # Î∞îÎ°ú Îã§Ïùå Î≤àÌò∏ÏôÄ ÎπÑÍµêÎßå Ìï¥ÎèÑ Ï†ëÎëêÏñ¥ Í¥ÄÍ≥ÑÎ•º Î™®Îëê Í≤ÄÏ∂ú Í∞ÄÎä•
        if phone_book[i+1].startswith(phone_book[i]):
            return False

    return True
````

````python
print(solution(a))  # False
print(solution(b))  # True  
print(solution(c))  # False
print(solution(d))  # False
````
