# 공백 기준으로 출력
https://www.codetree.ai/ko/trails/complete/curated-cards/test-output-based-on-space/description

```python
a = list(input().split())
b = list(input().split())

print(''.join(a)+''.join(b))
```

```python
a = list(input().split())
b = list(input().split())

print(*a, *b, sep='')
```

```python
# 공백을 포함한 문자열을 입력받습니다.
a = input()
b = input()

# 문자열을 전부 순회하며 공백을 제외한 모든 문자를 출력합니다.
for elem in a:
	if elem != " ":
		print(elem, end="")

for elem in b:
	if elem != " ":
		print(elem, end="")
```
